```python
# Question 2: Methods and Object Behavior of car (TASK 2)
class Car:
    def __init__(self, brand, model, year, mileage):
        self.brand = brand
        self.model = model
        self.year = year
        self.mileage = mileage
    def display_info(self):
        print(f"Car Information:\nBrand: {self.brand}\nModel: {self.model}\nYear: {self.year}\nMileage: {self.mileage}\n")

    def drive(self, distance):
        self.mileage += distance
        print(f"After driving {distance} km...")

car1 = Car("Toyota", "Corolla", 2020, 15000)# Creating a new Car object

# portraying the original information, driving the car, and then displaying updated information
car1.display_info()
car1.drive(100)
car1.display_info()

```

```python
# Question 3: Benefits of OOP (TASK 3)
class ElectricCar(Car):
    def __init__(self, brand, model, year, mileage, battery_capacity, charge_level):
        # Initializes an ElectricCar object with arguments like.
        #         brand (str): The brand of the car.
        #         model (str): The model of the car.
        #         year (int): The year of the car.
        #         mileage (int): The mileage of the car.
        #         battery_capacity (int): The battery capacity of the car in kWh.
        #         charge_level (int): The initial charge level of the car in percentage.
        super().__init__(brand, model, year, mileage)
        self.battery_capacity = battery_capacity
        self.charge_level = charge_level

    def charge(self, amount):
        #  Charges the car by a specified amount.
        # Args:
        #     amount (int): The amount to charge the car in percentage.
        # Returns:
        #     None
        
        self.charge_level = min(self.charge_level + amount, 100)
        print(f"Charging... New charge level: {self.charge_level}%")

# Creating an ElectricCar object
electric_car = ElectricCar("Rivian", "Type 3", 2022, 10000, 75, 50)

# Displaying the ElectricCar's information and charging it
electric_car.display_info()
electric_car.charge(30)

# Explanation of benefits of OOP
# The ElectricCar class demonstrates the principles of modularity, reusability, and maintainability
# by inheriting from the Car class and adding new attributes and methods specific to electric cars.
# This allows us to reuse the common attributes and methods of the Car class while adding new
# functionality specific to electric cars. 


```

```python
# Task 4: Comparison of OOP and Procedural Programming (TASK 4)

# Procedural implementation of Car functionality
def create_car(brand, model, year, mileage):
    car_info = {"brand": brand, "model": model, "year": year, "mileage": mileage}
    return car_info
def display_car_info(car_info):
    print("Car Information:")
    print(f"Brand: {car_info['brand']}")
    print(f"Model: {car_info['model']}")
    print(f"Year: {car_info['year']}")
    print(f"Mileage: {car_info['mileage']}")
def drive(car_info, distance):
    car_info["mileage"] += distance

car_info = create_car("Toyota", "Corolla", 2020, 15000)# Create a car using the procedural implementation


display_car_info(car_info)# Displaying the car information


drive(car_info, 100)# Drive the car and update the mileage


print("After driving 100 km...")# Displaying the updated car information
display_car_info(car_info)

# Answers to Task 4b questions
# The main differences between the OOP and procedural implementations are:
# - OOP uses classes and objects to organize code, while procedural programming uses functions and data structures.
# - OOP provides encapsulation, inheritance, and polymorphism, which are not present in procedural programming.
# - OOP is more modular, reusable, and maintainable than procedural programming.
```

```python

# Question 5: StudentManagement class TASK 5
# StudentManagement class with methods for adding, updating, and deleting students
class StudentManagement:
    def __init__(self):
        # Initializing an empty dictionary to store student data
        self.students = {}

    def add_student(self, student_id, name, age, course):
        # Adding a new student, handling if the student already exists
        if student_id in self.students:
            raise Exception(f"Student with ID {student_id} already exists.")
        else:
            self.students[student_id] = {"name": name, "age": age, "course": course}
            print(f"Student {name} added successfully.")

    def update_student(self, student_id, name=None, age=None, course=None):
        # Updating student details, handling if the student doesn't exist
        if student_id not in self.students:
            raise Exception(f"Student with ID {student_id} does not exist.")
        else:
            if name:
                self.students[student_id]["name"] = name
            if age:
                self.students[student_id]["age"] = age
            if course:
                self.students[student_id]["course"] = course
            print(f"Student {student_id}'s information updated successfully.")

    def delete_student(self, student_id):
        # Deleting a student, handling if the student doesn't exist
        if student_id not in self.students:
            raise Exception(f"Student with ID {student_id} does not exist.")
        else:
            del self.students[student_id]
            print(f"Student {student_id} deleted successfully.")

    def display_students(self):
        # Displaying all students
        if not self.students:
            print("No students available.")
        else:
            for student_id, info in self.students.items():
                print(f"Student ID: {student_id}, Name: {info['name']}, Age: {info['age']}, Course: {info['course']}")


# Example usage of the StudentManagement class
try:
    student_manager = StudentManagement()

    # Adding students
    student_manager.add_student(1, "Calvin", 20, "Computer Science")
    student_manager.add_student(2, "Ann", 22, "Engineering")
    
    # Attempting to add a student with an existing ID
    # student_manager.add_student(1, "Charlie", 21, "Mathematics")

    # Updating a student's information
    student_manager.update_student(1, age=21)

    # Displaying all students
    student_manager.display_students()

    # Deleting a student
    student_manager.delete_student(2)

    # Attempting to delete a non-existing student
    # student_manager.delete_student(3)

    # Displaying all students after deletion
    student_manager.display_students()

except Exception as e:
    print(e)
```
